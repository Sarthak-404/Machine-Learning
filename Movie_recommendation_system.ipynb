```python
import pandas as pd
import numpy as np
m = pd.read_csv(r'https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Movies%20Recommendation.csv')
m.head()
```

```python
m.info()
```

```python
m.columns
```

```python
m_features = m[['Movie_Genre','Movie_Keywords','Movie_Tagline', 'Movie_Cast','Movie_Director']].fillna(' ')
m_features.shape
```

```python
m_features
```

```python
x = m_features['Movie_Genre'] + ' ' + m_features['Movie_Keywords'] + ' ' + m_features['Movie_Tagline'] + ' ' + m_features['Movie_Cast'] + ' ' + m_features['Movie_Director']
x.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid = TfidfVectorizer()
X = tfid.fit_transform(x)
X.shape
```

```python
print(X)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
Similarity_Score = cosine_similarity(X)
Similarity_Score
```

```python
Similarity_Score.shape
```

```python
Favorite_Movie_Name = input(' Enter your favorite movie name : ')
All_Movies_Title_List = m['Movie_Title'].tolist()
import difflib
Movie_Recommendation = difflib.get_close_matches(Favorite_Movie_Name, All_Movies_Title_List)
print(Movie_Recommendation)
```

```python
Close_Match = Movie_Recommendation[0]
print(Close_Match)
```

```python
Index_of_Close_Match_Movie = m[m.Movie_Title == Close_Match]['Movie_ID'].values[0]
print(Index_of_Close_Match_Movie)
```

```python
Recommendation_Score = list(enumerate(Similarity_Score[Index_of_Close_Match_Movie]))
print(Recommendation_Score)
```

```python
len(Recommendation_Score)
```

```python
Sorted_Similar_Movies = sorted(Recommendation_Score, key = lambda x:x[1], reverse = True)
print(Sorted_Similar_Movies)
print('Top 30 Movies Suggested for you: \n')

i = 1

for movie in Sorted_Similar_Movies:
    index = movie[0]
    title_from_index = m[m.index==index]['Movie_Title'].values[0]
    if (i<31):
        print(i, '.',title_from_index)
        i+=1
```

```python
Movie_Name = input('Enter your favorite movie name: ')

list_of_all_titles = m['Movie_Title'].tolist()

Find_Close_Match = difflib.get_close_matches(Movie_Name, list_of_all_titles)

Close_Match = Find_Close_Match[0]

Index_of_Movie = m[m.Movie_Title == Close_Match]['Movie_ID'].values[0]

Recommendation_Score = list(enumerate(Similarity_Score[Index_of_Movie]))

sorted_similar_movies = sorted(Recommendation_Score, key = lambda x:x[1], reverse = True)

print('Top 10 Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = m[m.Movie_ID==index]['Movie_Title'].values
    if (i<11):
        print(i,'.',title_from_index)
        i+=1
```
